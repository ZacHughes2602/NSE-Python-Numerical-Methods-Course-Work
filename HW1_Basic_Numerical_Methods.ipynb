When a object is not at rest, its mass is increased based on the formula $$m = γ * m_0$$ where
$m$ is the mass, $m_0$ is the rest mass, and $γ$ is a relativistic factor given by $$γ = \frac{1}{\sqrt{1-\frac{v^2}{c^2}}}$$

The rest mass of a baseball is 145 g and the speed of light, c, is 2.99792458 × 10^8 m/s.
What is the mass of a baseball when thrown at v = 169.1 km/h (the fastest recorded
pitch)? How fast does it have to move to have a mass of 1.45 kg?

```python
import numpy as np


#Contants

m_0 = 145 #rest mass in grams
c = 2.99792458*10e8 #speed of light in m/s

#Question Parameters

v_kmh = 169/1 #velocity of ball in km/h
v = v_kmh * 1000/3600 #velocity converted to m/s
m_rel = 1450 #mass of ball in grams

#Equations

gamma = 1 / np.sqrt (1 - (v**2) / (c**2)) #Relavistic factor
m = gamma*m_0 #Mass increase formula


#Mass of baseball when thrown at 169.1 km/h


rf_1 = 1 / np.sqrt (1 - (v**2) / (c**2))
m_1 = rf_1*m_0


#Speed of baseball needed to have mass of 1.45 kg
from sympy import symbols, Eq, solve, sqrt

v_symb = symbols ('v', real = True, positive = True)
gamma_symb = 1 / sqrt (1 - (v_symb**2) / (c**2))
eq = Eq(gamma_symb * m_0, m_rel)

sol = solve(eq, v_symb)
v_req = sol[0].evalf()






#Outputs
print (f"Mass of the baseball when thrown at {v_kmh} km/h is {m_1:.4} grams")
print (f"Speed required for the baseball to have a mass of 1450 grams is {v_req:.4} m/s")
```

The cubic equation $${x^3}+{3b^2}x = 2c^3$$
has a root given by $$x = d - \frac{b^2}{d}$$
where $$d^3 = c^3 + \sqrt{b^6 +c^6}$$
This method was developed by Thomas Harriot (1560–1621), who also introduced the less
than and greater than symbols.
Write a program that prompts the user for coefficients of a general cubic, $$Ax^3+Bx^2+Cx+D$$
and determines if the cubic can be solved via Harriot’s method (i.e., B = 0). If it can be solved
via Harriot’s method, then print the solution. Also, print the residual from the root, that is
the value you get when you plug each into the original equation. Make sure that you do not
divide by zero, and that your method can handle imaginary roots.

```python
import numpy as np
import cmath

A = float(input("Enter coefficient A:"))
B = float(input("Enter coefficient B:"))
C = float(input("Enter coefficient C:"))
D = float(input("Enter coefficient D:"))

def solve(A, B, C, D):
    if B != 0:
        print("Thic cubic cannot be solved via Harriot's Method. Please enter different values (B must equal 0.")
    elif A == 0:
        print("Coefficient A cannot be zero for a cubic equation.")
    else:
        b_sq = C / (3 * A)
        c_cu = -D / (2 * A)
        
        inner_sqrt = cmath.sqrt(b_sq**3 + c_cu**2)
        d_cubed = c_cu + inner_sqrt

        d = d_cubed**(1/3)

        #Calculate the root and prevent division by zero
        if d == 0:
            #Avoid dividing by zero
            root = 0
        else:
            root = d - (b_sq / d)

        #Calculate residual
        res = A * root**3 + B * root**2 + C * root + D

        print(f"Root: {root:.4f}")
        print(f"Residual from root: {res:.4f}")

solve(A, B, C, D)
```

2.2: 

If we list all the natural numbers below 10 that are multiples of 3 or 7, we get 3, 6, 7,
and 9. The sum of these multiples is 25. Find the sum of all the multiples of 3 or 7 below
10000.

```python
total = 0

for nm in range (1,10000):
    if nm % 3 == 0 or nm % 7 ==0:
        total += nm

print (f"The sum of multiple of 3 or 7 between 1 and 10000 is {total}")
```

The foot bone connected to the leg bone
The leg bone connected to the knee bone

The bi bone connected to the bi+1 bone

Write a Python code that uses a for loop and a list that prints out an entire verse where

b = ["foot", "leg", "knee", "thigh", "back", "neck", "head"]

Write a Python code that asks the user to input the masses (in amu) of two reactants in a
nuclear reaction and the two products of the reaction. The code will output the Q value of the
reaction in MeV. The user should be able to enter zero for the mass in case the reaction is a decay (i.e., has only
one reactant) or is a reaction that has only one product. Use 1 amu = 931.494061 MeV/c^2.

```python
r_1 = float(input("Enter a mass in amu for reactant 1"))
r_2 = float(input("Enter a mass in amu for reactant 2"))
p_1 = float(input("Enter a mass in amu for product 1"))
p_2 = float(input("Enter a mass in amu for product 2"))

Mass_Dif = (r_1 + r_2) - (p_1 + p_2)
Q_Val = Mass_Dif * 931.494061

print(f"The Q value for this reaction is {Q_Val:.6} MeV")
```
